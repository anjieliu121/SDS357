```python
all_data = {
    "charlotte": charlotte_data,
    "durham": durham_data,
    "fayetteville": fayetteville_data,
    "greensboro": greensboro_data,
    "raleigh": raleigh_data,
    "statewide": statewide_data,
    "winston_salem": winston_salem_data,
}
```

```python
# Display what columns does each dataframe possess
all_columns = sorted(
    set().union(*[df.columns for df in all_data.values()])
)

column_existence = pd.DataFrame(
    {
        name: [col in df.columns for col in all_columns]
        for name, df in all_data.items()
    },
    index=all_columns
)

column_existence
```

```python
# Display data type of each column in each dataframe
dtype_table = pd.DataFrame(index=all_columns)

for name, df in all_data.items():
    dtype_table[name] = df.dtypes.reindex(all_columns)

dtype_table
```
